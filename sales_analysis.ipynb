# Sales Analysis Notebook

This notebook loads sample sales data, prints summary metrics, and visualizes sales by region and product.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_sample.csv')
df.head()
```

```python
print('Total sales:', df['sales_amount'].sum())
print('
Sales by region:')
print(df.groupby('region')['sales_amount'].sum())

print('
Sales by product:')
print(df.groupby('product')['sales_amount'].sum())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df.groupby('region')['sales_amount'].sum().plot(kind='bar', ax=axes[0], title='Sales by Region')
df.groupby('product')['sales_amount'].sum().plot(kind='bar', ax=axes[1], title='Sales by Product')
plt.tight_layout()
plt.show()
```
